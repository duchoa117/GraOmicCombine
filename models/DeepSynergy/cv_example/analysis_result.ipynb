```python
def mse(y,f):
    mse = ((y - f)**2).mean(axis=0)
    return mse
def pearson(y,f):
    rp = np.corrcoef(y, f)[0,1]
    return rp
```

```python
import torch
import torch.nn as nn
```

```python
#batch, feature
torch.manual_seed(42)
xt = torch.Tensor([
    [0.12312, 0.123312, 0.534534],
    [0.12312, 0.123312, 0.534534],
    [0.12312, 0.123312, 0.534534],
])
x_1 = torch.randn(3,9)
x_2 = torch.stack([x_1[2,:], x_1[1, :], x_1[0,:]])
# print(x_1)
# print(x_2)
key_drug = nn.Linear(9, 9)
key_xt_op = nn.Linear(3, 3)
at_fc = nn.Linear((9*2+3), 1) #21 -> 1
leaky_relu = nn.LeakyReLU()

key_xt = key_xt_op(xt)
key_drug_1 = key_drug(x_1)
key_drug_2 = key_drug(x_2)

a_drug_1 = torch.exp(leaky_relu(at_fc(torch.cat((key_drug_1, key_xt, key_drug_2), 1))))
a_drug_2 = torch.exp(leaky_relu(at_fc(torch.cat((key_drug_2, key_xt, key_drug_1), 1))))
# a_drug_1 = (torch.cat((key_drug_1, key_xt, key_drug_2), 1)) #1x21
# a_drug_2 = (torch.cat((key_drug_2, key_xt, key_drug_1), 1))
total = a_drug_1+a_drug_2
a_drug_1 = torch.div(a_drug_1, total)
a_drug_2 = torch.div(a_drug_2, total)
x_1 = a_drug_1*x_1
x_2 = a_drug_2*x_2
x_drug_combine = x_1 + x_2
```
