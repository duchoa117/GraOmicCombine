```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
arr = os.listdir()
arr
```

```python
data_paths = [
 'all_data.csv',
 'train_dc.csv',
 'val_dc.csv',
 'test_dc.csv',
 'blind_all_test.csv',
 'blind_1_drug_cell_val.csv',
 'blind_cell_val.csv',
 'mix_val.csv',
 'blind_all_val.csv',
 'blind_2_drug_test.csv',
 'blind_cell_test.csv',
 'blind_2_drug_val.csv',
 'blind_1_drug_cell_test.csv',
 'blind_1_drug_test.csv',
 'mix_test.csv',
 'blind_1_drug_val.csv',
]
data_paths.sort()
```

```python
for data_path in data_paths:
  print("_____________{0}___________".format(data_path))
  drug = list(pd.read_csv(data_path)['Drug1']) + list(pd.read_csv(data_path)['Drug1'])
  cell = list(pd.read_csv(data_path)['Cell line'])
  drug_distinct = list(set(drug))
  cell_distinct = list(set(cell))
  print('- Length drug:', len(drug))
  # print('Drug:', drug)

  print('- Length distinct drug:', len(drug_distinct))
  # print('Distinct drug:', drug_distinct)

  print('- Length cell:', len(cell))
  # print('Cell:', cell)

  print('- Length distinct cell:', len(cell_distinct))
  # print('Distinct cell:', cell_distinct)
```

```python
results = []
for data_path in data_paths:
  r = {'dataset': data_path}
  r.update(pd.read_csv(data_path)['synergy'].describe().to_dict())
  results.append(r)
```

```python
pd.DataFrame(results)
```

```python
pd.DataFrame(results).to_excel("dataset_distribution.xlsx", index=False)
```
