```python
import pandas as pd 
import numpy as np
```

```python
blind_cell = pd.read_csv("result_blind_drug.csv")
```

```python
def rmse(y,f):
    rmse = np.sqrt(((y - f)**2).mean(axis=0))
    return rmse
def mse(y,f):
    mse = ((y - f)**2).mean(axis=0)
    return mse
def pearson(y,f):
    rp = np.corrcoef(y, f)[0,1]
    return rp
```

```python
y = blind_cell["synergy"]
p = blind_cell["predict"]
print(rmse(y, p), mse(y, p), pearson(y, p))
```
